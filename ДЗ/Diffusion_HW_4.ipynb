```python
from pathlib import Path
from tqdm.auto import tqdm
from transformers import LlavaNextProcessor, LlavaNextForConditionalGeneration
import torch
from PIL import Image
import pandas as pd
from huggingface_hub import login
from datasets import load_dataset
from google.colab import drive
drive.mount('/content/drive')

model_id = "llava-hf/llava-v1.6-mistral-7b-hf"
```

```python
processor = LlavaNextProcessor.from_pretrained(model_id)
```

```python
model = LlavaNextForConditionalGeneration.from_pretrained(
    model_id,
    torch_dtype=torch.float16,
    low_cpu_mem_usage=True,
    load_in_4bit=True
)
```

```python
images = [f'/content/drive/My Drive/Colab_Notebooks/Diffusion/pusheen_data/{str(i)}.png' for i in range(81)]
len(images)
```

```python
captions = {}

for i, image_path in enumerate(tqdm(images)):
    image = Image.open(image_path)

    # Define a chat histiry and use `apply_chat_template` to get correctly formatted prompt
    # Each value in "content" has to be a list of dicts with types ("text", "image")
    conversation = [
        {
        "role": "user",
        "content": [
            {"type": "text", "text": """Describe the image in one concise sentence, always starting with 'Sticker of funny_cat Pusheen'.
                                        Include key details like colors, actions, emotions, and surroundings. Keep the description under 75 tokens.
                                        Do not generate a description without 'Sticker of funny_cat Pusheen'. Do not copy any example.
                                        Example: 'Sticker of funny_cat Pusheen sitting on a red sofa, happily enjoying a cupcake.'"""},
            {"type": "image"},
            ],
        },
    ]

    prompt = processor.apply_chat_template(conversation, add_generation_prompt=True)

    inputs = processor(images=image, text=prompt, return_tensors="pt").to("cuda:0")

    # autoregressively complete prompt
    with torch.inference_mode():
        output = model.generate(**inputs, max_new_tokens=77)

    caption = processor.decode(output[0], skip_special_tokens=True)
    caption = caption.split("[/INST]")[1].strip()

    captions[i] = caption
```
