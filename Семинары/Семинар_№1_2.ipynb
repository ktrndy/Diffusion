### Обозначения и важные величины

- Заглавная  $X$ обозначает случайную величину.
  
- Заглавная  $P(X)$ обозначает распределение вероятностей по этой переменной.
  
- Строчная  $x∼P(X)$ обозначает значение $x$, выбранное (∼) из распределения вероятностей $P(X)$ с помощью какого-либо генеративного процесса.
  
- Строчная $p(X)$ — это функция плотности распределения $X$. Это скалярная функция на пространстве меры $X$.
  

- $p(X=x)$ (сокращенно $p(x)$) обозначает функцию плотности, оцененную при конкретном значении $x$.

- Математическое ожидание по $p(x)$ : $\mathbb{E}_{p(x)}[f(x)] = \int f(x) p(x) dx.$


## Энтропия



Рассмотрим исходное определение энтропии, связанное с компрессией.

Лучшая компрессия для символа с вероятностью $p(x)$ — это код длины $L = \log (1/p)$. Об $L$ можно думать как о колличестве бит, чтобы передать информацию  (Шеннон и Уивер, 1949).

Чем реже встречается символ, тем более длинный код мы можем использовать, чтобы его закодировать.

Таким образом, количество информации можно рассматривать как среднюю длину всех кодов при идеальном сжатии или среднее колличество бит (если используем $\log_2$).

\begin{align}
                  \cssId{show-line-1}H[P] &= \mathbb{E}[\log\frac{1}{p(x)}] \text{ средняя длинна}  \\
                  &=\cssId{show-line-2}{\mathbb{E}[\log 1 - \log p(x)]} \\
                  &=\cssId{show-line-3}{\mathbb{E}[0 - \log p(x)]} \\
                  &=\cssId{show-line-4}{\mathbb{E}[- \log p(x)]} \\
                  &=\cssId{show-line-5}{\sum_{x\in\mathcal{X}}p(x) (- \log p(x))} \\
                  \cssId{show-line-6} H[P] &= -\sum_{x\in\mathcal{X}}p(x) \log p(x)  \\
                \end{align}

## Энтропия равномерного распределения




Возьмем определение энтропии, которое мы нашли выше.





\begin{align}
                  \cssId{show-line-7}H[P] &= -\sum_{x\in\mathcal{X}} p(x) \log p(x) \\
                  \cssId{show-line-8}u(x) &= \frac{1}{|\mathcal{X}|} \\
                  \cssId{show-line-9}U &: \forall x \in \mathcal{X}\quad p(x) = \frac{1}{|\mathcal{X}|}
                \end{align}

В равномерном распределении вероятность для каждого символа равна 1
деленной на количество возможных символов.

Для равномерного распределения все символы имеют одинаковую вероятность $u(x) = 1/|X|$.


Подставим эти значения:

\begin{align}
                  \cssId{show-line-10}H[U] &= -\sum_{x\in\mathcal{X}} u(x) \log u(x) \\
                  \cssId{show-line-11}H[U] &= -\sum_{x\in\mathcal{X}} \frac{1}{|\mathcal{X}|} \log \frac{1}{|\mathcal{X}|} \\
                  &= \cssId{show-line-12}{ -\sum_{x\in\mathcal{X}} \frac{1}{|\mathcal{X}|} (\log 1 - \log |\mathcal{X}|) }\\
                  &= \cssId{show-line-13}{ -\sum_{x\in\mathcal{X}} \frac{1}{|\mathcal{X}|}\log(1) - \frac{1}{|\mathcal{X}|}\log |\mathcal{X}| }\\
                  &= \cssId{show-line-14}{ -\sum_{x\in\mathcal{X}} \frac{1}{|\mathcal{X}|} * 0 - \frac{1}{|\mathcal{X}|}\log |\mathcal{X}| }\\
                  &= \cssId{show-line-15}{ -\sum_{x\in\mathcal{X}} -\frac{1}{|\mathcal{X}|}\log |\mathcal{X}| }\\
                  &= \cssId{show-line-16}{ \sum_{x\in\mathcal{X}} \frac{1}{|\mathcal{X}|}\log |\mathcal{X}| }\\
                  &= \cssId{show-line-17}{ \frac{1}{|\mathcal{X}|}\sum_{x\in\mathcal{X}} \log |\mathcal{X}| }\\
                  &= \cssId{show-line-18}{ \frac{1}{|\mathcal{X}|} * |\mathcal{X}| * \log |\mathcal{X}| }\\
                  \cssId{show-line-19}H[U] &= \log |\mathcal{X}|  \\
                \end{align}

Таким образом, энтропия равномерного распределения равна размеру **словаря**:  чем больше у нас слов, тем больше неопределенность того, какое слово мы вытащим следующим.

## Дивергенция Кульбака—Лейблера


Дивергенция Кульбака—Лейблера — это небольшая модификация нашей формулы для энтропии.

Допустим, мы будем **кодировать** распределение $P$ с помощью распределения $Q$. При этом $P$ и $Q$ не совпадают.

Мы знаем, что кодировать $P$ c помощью $P$ более разумно, так мы получим более правильные длинны для наших кодов. А если мы будем кодировать с $Q$, то нам потребуется больше бит.

Сколько бит информации нам понадобится в среднем, чтобы закодировать $P$ с помощью $Q$?

$L_p = \sum_{i=1}^{N} p(x_i) \cdot \log p(x_i) $  — среднее число бит, если кодируем с помощью $P(x)$.

$L_q = \sum_{i=1}^{N} p(x_i) \cdot \log q(x_i) $  — среднее число бит, если кодируем с помощью $Q(x)$.


Средняя разница в количестве бит:

$$ L_p -L_q =  \sum_{i=1}^{N} p(x_i) \cdot \log p(x_i) - \sum_{i=1}^{N} p(x_i) \cdot \log q(x_i) = \\  \sum_{i=1}^{N} p(x_i) \cdot (\log p(x_i) - \log q(x_i)) = \\
\sum_{i=1}^{N} p(x_i) \cdot \log \frac{ p(x_i)}{q(x_i)}.
$$

Собственно, это и есть дивергенция Кульбака—Лейблера:

$$ \sum_{i=1}^{N} p(x_i) \cdot \log \frac{ p(x_i)}{q(x_i)}. $$

## Дивергенция Кульбака—Лейблера между равномерным и некоторым распределением

Попробуем закодировать что-то с помощью равномерного распределения. Что тогда произойдет?

\begin{align}
                  \cssId{show-line-25}KL(P || U) &= \sum_{x\in\mathcal{X}} p(x) \log\frac{p(x)}{u(x)} \\
                  &= \cssId{show-line-26}{\sum_{x\in\mathcal{X}} p(x) (\log p(x) - \log u(x))} \\
                  &= \cssId{show-line-27}{\sum_{x\in\mathcal{X}} p(x)\log p(x) - p(x)\log u(x)} \\
                  &= \cssId{show-line-28}{\sum_{x\in\mathcal{X}} p(x)\log p(x) - \sum_{x\in\mathcal{X}} p(x)\log u(x)} \\
                  &= \cssId{show-line-29}{-H[P] - \sum_{x\in\mathcal{X}} p(x)\log u(x)} \\
                  &= \cssId{show-line-30}{-H[P] - \sum_{x\in\mathcal{X}} p(x)\log \frac{1}{|\mathcal{X}|}} \\
                  &= \cssId{show-line-31}{-H[P] - \sum_{x\in\mathcal{X}} p(x)(\log 1 - \log |\mathcal{X}|)} \\
                  &= \cssId{show-line-32}{-H[P] - \sum_{x\in\mathcal{X}} p(x)\log 1 - p(x)\log |\mathcal{X}|} \\
                  &= \cssId{show-line-33}{-H[P] - \sum_{x\in\mathcal{X}} p(x) * 0 - p(x)\log |\mathcal{X}|} \\
                  &= \cssId{show-line-34}{-H[P] - \sum_{x\in\mathcal{X}} - p(x)\log |\mathcal{X}|} \\
                  &= \cssId{show-line-35}{-H[P] + \sum_{x\in\mathcal{X}} p(x)\log |\mathcal{X}|} \\
                  &= \cssId{show-line-36}{-H[P] + \log |\mathcal{X}| \sum_{x\in\mathcal{X}} p(x)} \\
                  &= \cssId{show-line-37}{-H[P] + \log |\mathcal{X}| * 1} \\
                  &= \cssId{show-line-38}{-H[P] + \log |\mathcal{X}|} \\
                  &= \cssId{show-line-39}{\log |\mathcal{X}| -H[P]} \\
                  \cssId{show-line-40}KL(P || U) &= H[U] - H[P]  \\
                \end{align}
                
            
Таким образом, дивергенция между некоторым распределением и равномерным равна энтропии равномерного минус энтропия данного распределения.

### Прямой и обратный KL

Дивергенция Кульбака—Лейблера не является симметричной функцией расстояния, то есть $KL(P \parallel Q) \neq KL(Q \parallel P)$ (за исключением случая, когда $Q \equiv P$.

Рассмотрим сначала прямую KL-дивергенцию. Как мы видели из вышеприведенных выводов, мы можем записать KL как математическое ожидание "функции штрафа" $\log \frac{p(z)}{q(z)}$ по взвешивающей функции $p(z)$:

$KL(P \parallel Q) = \sum_z p(z) \log \frac{p(z)}{q(z)} = E_{p(z)} \left[ \log \frac{p(z)}{q(z)} \right].$

Заметим, что если в некоторой области $p(z)>0$, но $q(z)=0,$ то в этой области мы получим бесконечно большой штраф, $\lim_{q(z) \to 0} \log \frac{p(z)}{q(z)} \to \infty $. Это означает, что прямая KL-дивергенция будет большой там, где $Q(Z)$ не "покрывает" $P(Z)$.

Таким образом, прямая KL-дивергенция минимизируется, когда мы обеспечиваем условие $q(z) > 0$ там, где $p(z) > 0$.

Минимизация обратной KL-дивергенции имеет другое поведение:

$KL(Q \parallel P) = \sum_z q(z) \log \frac{q(z)}{p(z)} = E_{q(z)} \left[ \log \frac{q(z)}{p(z)} \right]$

Мы должны обеспечить, чтобы  $q(Z) = 0$ там, где знаменатель $p(Z) = 0$, иначе KL-дивергенция будет стремиться к бесконечности.

### Еще один способ посмотреть на KL-дивиргенцию


$$KL(P \parallel Q)  = \sum_{i=1}^{N} p(x_i) \cdot \log p(x_i) - \sum_{i=1}^{N} p(x_i) \cdot \log q(x_i) =\\
=\text{Entropy} - \sum_{i=1}^{N} p(x_i) \cdot \log q(x_i)= \\
=\text{Entropy} - \sum_{i=1}^{N} p(x_i) \cdot \log q(x_i) = \text{Entropy} - \text{CrossEntropy}$$

А еще вспомним, что кросс-энтропия и логарифм правдоподобия в задаче классификации — это одно и тоже.

### Пример с кодом

Возьмем в качестве $ p(x)$ смесь двух одномерных гауссовских распределений:

$$p(x) = \pi N(\mu_1, \sigma_1^2) + (1 - \pi) N(\mu_2, \sigma_2^2).$$

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.distributions as dist
import matplotlib.pyplot as plt
import numpy as np
```

```python
def gen_data(mu_1=0, sigma_1=1, mu_2=0, sigma_2=1, proportion=0.3, n=1000):
    data = np.zeros(n)
    for i in range(n):
        if np.random.rand() < proportion:
            data[i] = np.random.normal(loc=mu_1, scale=sigma_1)
        else:
            data[i] = np.random.normal(loc=mu_2, scale=sigma_2)
    return data


data = gen_data(2, 1, 6, 0.5, 0.3);

# Plot the sampled data
plt.figure(figsize=(12, 6))

# Histogram of the sampled data
plt.hist(data, bins=100, density=True, alpha=0.6, color='g', label='Sampled Data');
```

```python
class Model(nn.Module):
    def __init__(self, mean1, mean2, cov):
        super(Model, self).__init__()
        self.mean1 = torch.tensor(mean1, dtype=torch.float32)
        self.mean2 = torch.tensor(mean2, dtype=torch.float32)
        self.cov = torch.tensor(cov, dtype=torch.float32)
        self.d = self.cov.shape[0]

    def log_density(self, zs):
        clust1_density = 0.5 * dist.MultivariateNormal(self.mean1, self.cov).log_prob(zs)
        clust2_density = 0.5 * dist.MultivariateNormal(self.mean2, self.cov).log_prob(zs)
        return torch.logsumexp(torch.stack([clust1_density, clust2_density]), dim=0)

    def sample(self, n):
        num_clust1 = torch.binomial(torch.tensor(n, dtype=torch.float32), torch.tensor(0.5))
        num_clust2 = n - num_clust1
        samples_mode1 = dist.MultivariateNormal(self.mean1, self.cov).sample((num_clust1,))
        samples_mode2 = dist.MultivariateNormal(self.mean2, self.cov).sample((num_clust2,))
        samples = torch.cat([samples_mode1, samples_mode2], dim=0)
        return samples

class ApproxMFGaussian(nn.Module):
    def __init__(self):
        super(ApproxMFGaussian, self).__init__()

    def log_density(self, var_param, z):
        mean = var_param[:2]
        log_std = var_param[2:]
        std = torch.exp(log_std)
        return dist.MultivariateNormal(mean, torch.diag(std)).log_prob(z)

    def sample(self, var_param, S):
        mean = var_param[:2]
        log_std = var_param[2:]
        std = torch.exp(log_std)
        return mean + std * torch.randn(S, 2)

    def entropy(self, var_param):
        log_std = var_param[2:]
        return 0.5 * 2 * (1.0 + np.log(2 * np.pi)) + torch.sum(log_std * 2)

d = 2

# Generate data
mean1 = np.array([-4, -4])
mean2 = np.array([4, 4])
cov = np.eye(d)

model = Model(mean1=mean1, mean2=mean2, cov=cov)
approx = ApproxMFGaussian()

S = 1000

# Variational parameters: first d elements are the mean, last d elements are the diagonal of log-covariance
variational_mean = np.array([0, 0])
variational_log_cov_diag = np.ones(d)
variational_param = np.concatenate([variational_mean, variational_log_cov_diag])
variational_param = torch.tensor(variational_param, requires_grad=True)

def objective(variational_param):
    samples = approx.sample(variational_param, S)
    lik = torch.mean(model.log_density(samples))
    entropy = approx.entropy(variational_param)
    elbo = lik + entropy
    return -elbo / S

def plot_isocontours(ax, func, xlimits=[-10, 10], ylimits=[-10, 10], numticks=101, cmap='viridis', label=''):
    x = np.linspace(*xlimits, num=numticks)
    y = np.linspace(*ylimits, num=numticks)
    X, Y = np.meshgrid(x, y)
    zs = func(torch.tensor(np.concatenate([np.atleast_2d(X.ravel()), np.atleast_2d(Y.ravel())]).T, dtype=torch.float32))
    Z = zs.reshape(X.shape).detach().numpy()
    plt.contour(X, Y, Z, cmap=cmap, label=label)
    ax.set_yticks([])
    ax.set_xticks([])


imgs = []

def print_perf(params, iter):
    bound = objective(params)
    plt.cla()
    target_distribution = lambda x : torch.exp(model.log_density(x))
    plot_isocontours(ax, target_distribution, label="p(z)")

    variational_contour = lambda x: dist.MultivariateNormal(params[:2], torch.diag(torch.exp(params[2:]))).log_prob(x).exp()
    plot_isocontours(ax, variational_contour, cmap='plasma', label="q(z)")

    plt.draw()
    plt.pause(0.01)

num_epochs = 100
step_size = 0.1

optimizer = optim.Adam([variational_param], lr=step_size)


fig, ax = plt.subplots()

for epoch in range(num_epochs):
    optimizer.zero_grad()
    loss = objective(variational_param)
    loss.backward()
    optimizer.step()
    if epoch%10==0:
        print_perf(variational_param, epoch)
```
